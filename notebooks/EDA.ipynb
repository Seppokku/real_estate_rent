#### Удалим столбцы 'Unnamed: 0' и 'Id объявления', т.к. это уникальные идентификаторы объектов

```python
data.drop(['Unnamed: 0', 'ID  объявления'], axis=1, inplace=True)
```

#### Возьмем колонки до столбца 'ремонт' (так поделились) и назовем колонки по английски

```python
df = data[['Количество комнат', 'Тип', 'Метро', 'Адрес', 'Площадь, м2', 'Дом', 'Парковка', 'Цена', 'Телефоны', 'Описание']]
```

```python
df.columns = ['rooms', 'type', 'subway', 'adress', 'square', 'building', 'parking', 'price', 'contact_phone', 'description']
```

#### Удалим стобец 'phone', т.к. это контактная информация риелтора/владельца объекта недвижимости

```python
df.drop('contact_phone', axis=1, inplace=True)
```

```python
df.info()
```

#### Посмотрим пропущенные значение в колонке 'rooms' и разберемся в них

```python
df[df['rooms'].isna()]
```

#### Квартиры с пропусками в колонке 'rooms' это студии, у которых не указывается количество комнат в объявлении, можем пока его заменить значением - 0 

```python
df['rooms'] = df['rooms'].fillna(0)
```
